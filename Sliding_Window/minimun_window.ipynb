Given two strings `s` and `t` of lengths `m` and `n` respectively, return the minimum window substring of `s` such that every character in `t` (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The test cases will be generated such that the answer is unique.

**Example 1:**

Input: s = "ADOBECODEBANC", t = "ABC"
Output: "BANC"
Explanation: The minimum window substring "BANC" includes 'A', 'B', and 'C' from string t.

**Example 2:**

Input: s = "a", t = "a"
Output: "a"
Explanation: The entire string s is the minimum window.

**Example 3:**
Input: s = "a", t = "aa"
Output: ""
Explanation: Both 'a's from t must be included in the window. Since the largest window of s only has one 'a', return empty string.


```python

class Solution:
    def minWindow(self, s: str, t: str) -> str:
        # Valido se a string é vazia, se for retorno vazio
        if t == '': return ''
        
        # Gero dois hashmaps para contagem de frequencia dos elementos
        countT, window = {}, {}
        
        # Contabilizo a frequencia do countT
        for c in t:
            countT[c] = 1 + countT.get(c, 0)
        
        # Inicializando variáveis core
        have, need = 0, len(countT)
        res, resLen = [-1, -1], float("infinity")
        l = 0
        
        # Itero pela string principal
        for r in range(len(s)):
            # pego o caractere atual
            c = s[r]
            
            # preencho a frequencia do caracter no hashmap
            window[c] = 1 + window.get(c, 0)
            
            # caso esse caractere seja um dos presentes em T
            # e tenha a frequência necessária, incrementa no have
            if c in countT and window[c] == countT[c]:
                have += 1
            
            # Enquando eu tiver o necessário eu tento diminuir a substring
            while have == need:
                # Verifico se o tamanho da substr atual é menor que a ultima
                # Se for atualiza
                if (r - l + 1) < resLen:
                    res = [l, r]
                    resLen = r - l + 1
                # Dou um pop no elemento mais a esquerda da minha window
                # Ou seja, removo 1 da frequencia de tal caracter
                window[s[l]] -= 1
                
                # Se o caracter removido é um dos T e agora eu não possuo
                # A frequencia necessária, diminui 1 em have
                if s[l] in countT and window[s[l]] < countT[s[l]]:
                    have -= 1
                
                # Atualizo o left pointer
                l += 1
        
        l, r = res
        return s[l:r+1] if resLen != float("infinity") else ''
```
