## Max Sliding Window

You are given an array of integers `nums`, and there is a sliding window of size `k` which is moving from the very left of the array to the very right. You can only see the `k` numbers in the window, and each time the sliding window moves right by one position.

### Task

Return the max sliding window.

### Example

#### Example 1:

```python
Input: nums = [1, 3, -1, -3, 5, 3, 6, 7], k = 3
Output: [3, 3, 5, 5, 6, 7]
Explanation:
Window position           Max
---------------          -----
[1, 3, -1] -3, 5, 3, 6, 7  3
1, [3, -1, -3] 5, 3, 6, 7   3
1, 3, [-1, -3, 5] 3, 6, 7   5
1, 3, -1, [-3, 5, 3] 6, 7   5
1, 3, -1, -3, [5, 3, 6] 7   6
1, 3, -1, -3, 5, [3, 6, 7]  7

Example 2:
Input: nums = [1], k = 1
Output: [1]

Constraints:
1 <= nums.length <= 105
-104 <= nums[i] <= 104
1 <= k <= nums.length


````python
import collections
from typing import List


class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        output = []
        deque = collections.deque() # Armazeno index
        l = r = 0
        
        # Itera todo o array
        while r < len(nums):
            # Enquanto estiver algo no deque e o último elemento adicionado for menor que o atual
            # Removo o elemento do deque, pois ele não é o maior elemento do range
            while deque and nums[deque[-1]] < nums[r]:
                deque.pop()
            
            # Se removi o ultimo elemento ou não tem nada no deque, appenda o elemento
            deque.append(r)
            
            # Removo o elemento a esquerda da window
            if l > deque[0]:
                deque.popleft()
                
            # Se meu r chegou no range máximo
            if r + 1 >= k:
                output.append([deque[0]])
                l += 1
            
            r += 1
                
        return output
        
````
